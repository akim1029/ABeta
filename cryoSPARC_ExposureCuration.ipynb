```python
import numpy as np
import pandas as pd
import sklearn.datasets
from scipy import stats
```

```python
# Import data generated by cryoSPARC
df = pd.read_csv("/Users/Ashley/Documents/pdnc.csv", header=None)
```

```python
# Convert dataframe to numerical array
numdat = np.zeros((2327,12))
for i in range(2326):
    for j in range(11):
        numdat[i,j] = pd.to_numeric(df.loc[i+1,j+1])
```

```python
# Generate thresholds for exposure curation
for k in range(11):
    min_thresh = np.max(numdat[:,k])
    max_thresh = np.min(numdat[:,k])
    z = stats.zscore(numdat[:,k])
    for l in range(2326):
        entry = numdat[l,k]
        if abs(z[l]) > 2:
            if entry < np.average(numdat[:,k]):
                if entry > max_thresh:
                    max_thresh = entry
            else: 
                if entry < min_thresh:
                    min_thresh = entry
    print(max_thresh, min_thresh)
                
```
