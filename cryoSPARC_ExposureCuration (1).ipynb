```python
import numpy as np
import pandas as pd
import sklearn.datasets
from scipy import stats
```

```python
# Import data generated by cryoSPARC

df = pd.read_csv("/Users/Ashley/Documents/pdnc.csv", header=None)
```

```python
# Convert dataframe to numerical array

numdat = np.zeros((2327,12))
for i in range(2326):
    for j in range(11):
        numdat[i,j] = pd.to_numeric(df.loc[i+1,j+1])
    
```

```python
# Generate thresholds for exposure curation

# Parameters: data (2D-array) and zthresh (cutoff Z-score for outlier)

def ThresholdGenerator(data, zthresh):
    
    rows = data.shape[0]
    cols = data.shape[1]
    
    for k in range(cols):
        
        min_thresh = np.min(data[:,k])
        max_thresh = np.max(data[:,k])
        z = stats.zscore(data[:,k])
        
        for l in range(rows):
            
            entry = data[l,k]
            if abs(z[l]) > zthresh:
                
                if entry < np.average(data[:,k]):
                    if entry > min_thresh:
                        min_thresh = entry
                else: 
                    if entry < max_thresh:
                        max_thresh = entry
                        
        print(str(min_thresh) + ' < ' + str(df.loc[0,k+1]) + ' < ' + str(max_thresh))
        
```

```python
ThresholdGenerator(numdat,2)
```
